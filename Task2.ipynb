# Data Cleaning

```python
columns_to_remove = ['SibSp', 'Parch', 'Ticket', 'Fare','Cabin','Embarked']
df = df.drop(columns=columns_to_remove, errors='ignore')
```

```python
print(df)
```

```python
df.head()
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
df.dropna(subset=['Age'], inplace=True)
```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
print("\nCleaned DataFrame:")
print(df)
```

# EDA(Exploratory Data Analysis):

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(df.describe())
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Count of Survived and Not Survived')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(x='Age', hue='Survived', data=df, fill=True, common_norm=False)
plt.title('Age Distribution by Survival Status')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, ci=None)
plt.title('Survival Rate by Sex')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Count of Survival by Sex')
plt.show()
```
